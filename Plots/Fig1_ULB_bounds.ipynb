```python
import matplotlib
#matplotlib.rc_file('matplotlibrc')
import matplotlib.pyplot as plt

# Setting fig size
plt.rcParams["figure.figsize"] = (40,5) 

# for Sgr A*
#a=0.94
ub=1.17e-17  # Upper bound
lb0=1.678e-18 # Lower bound Spin 0
lb1=3.512e-19 # Lower bound Spin 1
lb2=8.813e-19 # Lower bound Spin 2
#a=0.5
ubs=4.49e-18  # Upper bound
lb0s=1.859e-18 # Lower bound Spin 0
lb1s=3.877e-19 # Lower bound Spin 1
lb2s=1.043e-18 # Lower bound Spin 2

#for Ark 120
ubA=1.137e-19
lbA1=4.094e-20
#Unal et al
ubU=1.7e-17
lbU0=2e-20

# create dataset
#height = [3, 2, 1]
bars = ('Spin 0', 'Spin 2', 'Spin 1')
bottom = [2, 1, 0]
bottoms = [2, 1, 0]
# Choose the width of each bar and their positions
#a=0.94
width = [ub-lb0,ub-lb2,ub-lb1]
x_pos = [(ub+lb0)/2,(ub+lb2)/2,(ub+lb1)/2]
c = ['Green', 'Grey', 'Orange']
#a=0.5
widths = [ubs-lb0s,ubs-lb2s,ubs-lb1s]
x_poss = [(ubs+lb0s)/2,(ubs+lb2s)/2,(ubs+lb1s)/2]

# Make the plot
plt.bar(x_pos, 0.9,bottom=bottom, width=width, color=c,alpha=0.5)
plt.bar(x_poss, 0.9,bottom=bottoms, width=widths, color=c,alpha=1)
#plt.bar((ubU+lbU0)/2 ,1, width=ubU-lbU0,bottom=3.5, fill=False,linestyle='--')#, color=c)
plt.bar((ubA+lbA1)/2 ,1, width=ubA-lbA1,bottom= 0, linestyle='--', color='Red',alpha=0.5)
plt.bar((ubU+lbU0)/2 ,1, width=ubU-lbU0,bottom= 3.4, linestyle='--', color='Blue',alpha=0.1)

# Add Text Label
plt.text(4e-18,2.4,"S",ha='center',fontsize=20)
plt.text(2e-18,0.4,"V",ha='center',fontsize=20)
plt.text(4e-18,1.4,"T",ha='center',fontsize=20)
plt.text(6e-19,1.1,"Sagittarius A*",ha='center',fontsize=28)

plt.text(7e-20,0.4,"V",ha='center',fontsize=20)
plt.text(7e-20,1.1,"Ark 120",ha='center',fontsize=28)

plt.text(2e-19,3.76,"Unal et al",ha='center',fontsize=28)

plt.xlabel(r'$\mu$ (eV)', fontsize=28)

# Create names on the x-axis
plt.xticks(x_pos, bars)

# Removing y axis
plt.yticks([])

#get current axes
ax = plt.gca()

#hide y-axis
#ax.get_yaxis().set_visible(False)

plt.xscale("log")
plt.xlim(3e-20, 3e-17)

# Command used for hiding whitespaces and border.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.savefig('Denton.jpg', bbox_inches='tight', pad_inches=0)

# Show graphic
plt.show()
#print(ub)
```
